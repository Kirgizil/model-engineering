#### *Shapiro-Wilk Test*

```python
from scipy import stats

features = ['radius', 'texture', 'perimeter', 'area', 'smoothness', 'compactness', 'concavity', 'concave points', 'symmetry', 'fractal_dimension']

for feature in features:
    name = feature
    x1=name+'_mean'
    x2=name+'_se'
    x3=name+'_worst'

    alpha = 0.01

    statistic, p_value = stats.shapiro(df[df['diagnosis']=='B'][x1])
    if p_value > alpha:
        print("Benign {0} p-value:".format(x1), round(p_value, 2))
    statistic, p_value = stats.shapiro(df[df['diagnosis']=='B'][x2])
    if p_value > alpha:
        print("Benign {0} p-value:".format(x2), round(p_value, 2))
    statistic, p_value = stats.shapiro(df[df['diagnosis']=='B'][x3])
    if p_value > alpha:
        print("Benign {0} p-value:".format(x3), round(p_value, 2))

    statistic, p_value = stats.shapiro(df[df['diagnosis']=='M'][x1])
    if p_value > alpha:
        print("Malignant {0} p-value:".format(x1), round(p_value, 2))
    statistic, p_value = stats.shapiro(df[df['diagnosis']=='M'][x2])
    if p_value > alpha:
        print("Malignant {0} p-value:".format(x2), round(p_value, 2))
    statistic, p_value = stats.shapiro(df[df['diagnosis']=='M'][x3])
    if p_value > alpha:
        print("Malignant {0} p-value:".format(x3), round(p_value, 2))
```

#### *Welch's t-test for all available normalized features with a significance level of Alpha=0.01*
*Assumes normal distribution and no interaction between features. Gives a sense about which features are not significantly distinct*

```python
from scipy.stats import ttest_ind

features = ['radius', 'texture', 'perimeter', 'area', 'smoothness', 'compactness', 'concavity', 'concave points', 'symmetry', 'fractal_dimension']

for feature in features:
    name=feature
    x1=name+'_mean'
    x2=name+'_se'
    x3=name+'_worst'
    
    distribution1 = df_normalized[df_normalized['diagnosis'] == 'B'][x1].tolist()
    distribution2 = df_normalized[df_normalized['diagnosis'] == 'M'][x1].tolist()
    t_statistic, p_value = ttest_ind(distribution1, distribution2, equal_var=False)
    
#    print(x1 + ": T-statistic:", t_statistic)
#    print(x1 + ": P-value:", p_value)
    
    alpha = 0.01  # Significance level
    if p_value < alpha:
        pass
#        print(x1 + " is significantly different.")
    else:
        print(x1 + " is not significantly different.(p={0:.2f})".format(p_value))
    
    distribution1 = df_normalized[df_normalized['diagnosis'] == 'B'][x2].tolist()
    distribution2 = df_normalized[df_normalized['diagnosis'] == 'M'][x2].tolist()
    t_statistic, p_value = ttest_ind(distribution1, distribution2, equal_var=False)
    
#    print(x2 + ": T-statistic:", t_statistic)
#    print(x2 + ": P-value:", p_value)
    
    alpha = 0.01  # Significance level
    if p_value < alpha:
        pass
#        print(x2 + " is significantly different.")
    else:
        print(x2 + " is not significantly different.(p={0:.2f})".format(p_value))
    
    distribution1 = df_normalized[df_normalized['diagnosis'] == 'B'][x3].tolist()
    distribution2 = df_normalized[df_normalized['diagnosis'] == 'M'][x3].tolist()
    t_statistic, p_value = ttest_ind(distribution1, distribution2, equal_var=False)
    
#    print(x3 + ": T-statistic:", t_statistic)
#    print(x3 + ": P-value:", p_value)
    
    alpha = 0.01  # Significance level
    if p_value < alpha:
        pass
#        print(x3 + " is significantly different.")
    else:
        print(x3 + " is not significantly different. (p={0:.2f})".format(p_value))
```

#### *Calculate % overlapp of the Benign and Malignant Cell normal sq2SUM distribution*

```python
from scipy.integrate import quad
from scipy.stats import norm
import numpy as np

mu1, sigma1 = B_mean, B_std  # Mean and standard deviation of distribution 1
mu2, sigma2 = M_mean, M_std  # Mean and standard deviation of distribution 2

# Determine the range of overlap
lower_bound = max(mu1 - 3 * sigma1, mu2 - 3 * sigma2)
upper_bound = min(mu1 + 3 * sigma1, mu2 + 3 * sigma2)

# Calculate the area of overlap
overlap_area, _ = quad(lambda x: norm.pdf(x, mu1, sigma1) * norm.pdf(x, mu2, sigma2), lower_bound, upper_bound)

# Calculate the total area under both curves
total_area_B, _ = quad(lambda x: norm.pdf(x, mu1, sigma1), -np.inf, np.inf)
total_area_M, _ = quad(lambda x: norm.pdf(x, mu2, sigma2), -np.inf, np.inf)
total_area = total_area_B + total_area_M

# Calculate the percentage overlap
percentage_overlap = (overlap_area / total_area) * 100

print("Percentage overlap: {0:.2f}%".format(percentage_overlap))
```
